```python
f_p = open("data/resized_train_drug_facts_postive.txt", "w")
f_n = open("data/resized_train_drug_facts_negative.txt", "w")

jpg_files = glob.glob("/data/tmp/nutrition_label_dataset_abon_resized/toothpastes/US_toothpastes/*.jpeg")
jpg_files
```

```python
for jpeg in jpg_files:
    txt = re.sub(".jpeg",".txt", jpeg)
    if os.path.exists(txt):
        image = cv2.imread(jpeg)
        img_w = image.shape[1]
        img_h = image.shape[0]
        print(jpeg, img_w, img_h)
        ann_txt = open(txt)
        l_w = [jpeg]
        for t in ann_txt:
            t = t.rstrip()
            idx, xcen_n, ycen_n, w_n, h_n = t.split(" ")
            print(idx, xcen_n, ycen_n, w_n, h_n)
            x_cen =  img_w * float(xcen_n)
            y_cen = img_h * float(ycen_n)
            w = img_w * float(w_n)
            h = img_h * float(h_n)
            xmin = int(x_cen - (w/2) - BOUNDING_BOX_PADDING)
            ymin = int(y_cen - (h/2) - BOUNDING_BOX_PADDING)
            xmax = int(x_cen + (w/2) + BOUNDING_BOX_PADDING)
            ymax = int(y_cen + (h/2) + BOUNDING_BOX_PADDING)
            l_w.append("{},{},{},{},{}".format(xmin, ymin, xmax, ymax, 1))
        print(" ".join(l_w))
        f_p.write(" ".join(l_w) + "\n")
    else:
        f_n.write("{} 0,0,1,1,0\n".format(jpeg))
```

```python
f_p.close()
f_n.close()
```
