```python
import glob
import re
import os
import cv2

BOUNDING_BOX_PADDING = 5
```

```python
f_w = open("train_food_facts.txt", "w")
ann_files = glob.glob("/data/tmp/toothpastes/US_toothpastes/*.txt")
ann_files
```

```python
for ann in ann_files:
    jpeg = re.sub(".txt",".jpeg", ann)
    if os.path.exists(jpeg):
        image = cv2.imread(jpeg)
        img_w = image.shape[1]
        img_h = image.shape[0]
        print(jpeg, img_w, img_h)
        ann_txt = open(ann)
        l_w = [jpeg]
        for t in ann_txt:
            t = t.rstrip()
            idx, xcen_n, ycen_n, w_n, h_n = t.split(" ")
            print(idx, xcen_n, ycen_n, w_n, h_n)
            x_cen =  img_w * float(xcen_n)
            y_cen = img_h * float(ycen_n)
            w = img_w * float(w_n)
            h = img_h * float(h_n)
            xmin = int(x_cen - (w/2) - BOUNDING_BOX_PADDING)
            ymin = int(y_cen - (h/2) - BOUNDING_BOX_PADDING)
            xmax = int(x_cen + (w/2) + BOUNDING_BOX_PADDING)
            ymax = int(y_cen + (h/2) + BOUNDING_BOX_PADDING)
            l_w.append("{},{},{},{},{}".format(xmin, ymin, xmax, ymax, idx))
        print(" ".join(l_w))
        f_w.write(" ".join(l_w) + "\n")
```

```python
f_w.close()
```
